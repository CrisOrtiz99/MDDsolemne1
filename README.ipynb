### Análisis sobre datos de accidentes viales en USA 2015

El siguiente repositorio continen analisís y tratamiento de datos sobre accidentes viales en USA transurridos en 2015 para la solmene
numero 1 de la asignatura Mineria de datos.

### Requerimientos

La lista de librerias requeridas para ejecutar correctamente el analisis se encuentran en el archivo requirements.txt y son las sigueintes:

- matplotlib==3.6.3
- numpy==1.24.2
- pyparsing==3.0.9
- seaborn==0.12.2
- pandas==2.0.0
- google-cloud-bigquery==3.9.0
- python==3.11.2


# Estructura del repositorio
El repositorio esta diseñado con 8 comimts que contiene los codigos que se ocuparon en orden para el procedimiento.

Los resultado se encuentrar estructurados de la siguiente manera, la cual esta planteada en la solemne:
- Archivo csv que corresponde a el diccionario de datos del Data Frame
- Codigos del analisis graficos solicitado:
    - Numero de accidentes por estado (`state_name`)
    - numero de accidentes por uso de tierra (`land_use`)
    - numero de accidentes por empresa de carreteras (`ownership_name`)
    - numero de accidentes por carretera (`trafficway_identifier`)
    - análisis mensual de accidentes por estado
    - análisis según la hora del dia para los estados con mayor numero de muertes
    - la razón entre números de accidentes y conductores ebrios

## Conclusiones

A partir de los graficos obtenidos y la informacion analizada podemos concluir que los tres Estado con IDH problematico, la extencion de la loguitud de la carretera y la ubicacion de esta misma influyen en la cantidad de accidentes, dada esta infromacion se pueden generar planes de contingencia y de educacion para las regiones.
Ademas en conjunto a lo anterior pudimos observar una alta taza de accidentes por conductores ebrios, los cuales se traducen en una carencia de educacion vial en la zona y de controles de alcoholismo.
